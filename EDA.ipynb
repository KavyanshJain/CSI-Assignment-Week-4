# Country Specific Analysis and Trends

```python
df['Nationality'].value_counts()
```

USA has the highest count of athletes.

```python
nat_sport = df.groupby(['Nationality','Sport'])['Name'].count()
nat_sport.to_csv('nat_sport.csv')
nat_sport = pd.read_csv('nat_sport.csv')
nat_sport.rename(columns={'Name':'NameCount'},inplace=True)
nat_sport.head()
```

```python
fig,axes=plt.subplots(figsize=(13,13),nrows=2,ncols=1,dpi=100)

df['Nationality'].value_counts().plot(kind='bar',cmap='Set2',ax=axes[0])
axes[0].set_title('Numbar of Nationality')

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
nationality=df['Nationality'].value_counts()
axes[1].pie(nationality,labels=nationality.index,autopct='%.1f%%',colors=colors,shadow=True)
axes[1].set_title('Distribution of Nationality')

fig.tight_layout()
plt.show()
```

This bar graph and pie chart showcase us how are the highest earning Athletes are spreaded in their nationality.
USA has the highest share of high earning Athletes, followed by UK and Germany in 2nd and 3rd place respectiviely.

```python
fig,axes=plt.subplots(figsize=(13,13),nrows=2,ncols=1,dpi=100)

df['Sport'].value_counts().plot(kind='bar',cmap='Set2',ax=axes[0])
axes[0].set_title('Numbar of Sport')

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
sport=df['Sport'].value_counts()
axes[1].pie(sport,labels=sport.index,autopct='%.1f%%',colors=colors,shadow=True)
axes[1].set_title('Distribution of Sport')

fig.tight_layout()
plt.show()
```

```python
fig=px.scatter(nat_sport,x='Nationality',y='NameCount',color='Sport',size='NameCount',title='Unique number of Athletes in each country')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
px.bar(df,x='Year',y='earnings ($ million)',color='Nationality',title='Earnings trend')
```

This chart shows the how much share is of different nations.

Let's check the Country wise trends of share throughout the decades.

```python
nat_year = df.groupby(['Nationality','Year',])['earnings ($ million)'].sum()
nat_year.to_csv('nat_year.csv')
nat_year = pd.read_csv('nat_year.csv',index_col=False)
nat_year.tail()
```

```python
nat_year1990=nat_year.loc[nat_year['Year']==1990]
nat_year1990=pd.melt(nat_year1990,id_vars=['Nationality'],value_vars='earnings ($ million)',value_name='Value')
nat_year1990['percent']=[round(i*100/sum(nat_year1990.Value),1) for i in nat_year1990.Value]
nat_year2000=nat_year.loc[nat_year['Year']==2000]
nat_year2000=pd.melt(nat_year2000,id_vars=['Nationality'],value_vars='earnings ($ million)',value_name='Value')
nat_year2000['percent']=[round(i*100/sum(nat_year2000.Value),1) for i in nat_year2000.Value]
nat_year2010=nat_year.loc[nat_year['Year']==2010]
nat_year2010=pd.melt(nat_year2010,id_vars=['Nationality'],value_vars='earnings ($ million)',value_name='Value')
nat_year2010['percent']=[round(i*100/sum(nat_year2010.Value),1) for i in nat_year2010.Value]
nat_year2020=nat_year.loc[nat_year['Year']==2020]
nat_year2020=pd.melt(nat_year2020,id_vars=['Nationality'],value_vars='earnings ($ million)',value_name='Value')
nat_year2020['percent']=[round(i*100/sum(nat_year2020.Value),1) for i in nat_year2020.Value]

fig,axes=plt.subplots(figsize=(20,5),nrows=1,ncols=4)

axes[0].pie(nat_year1990.Value,labels=nat_year1990.Nationality,autopct='%1.1f%%',pctdistance=0.9)
axes[0].set_title('1990')

axes[1].pie(nat_year2000.Value,labels=nat_year2000.Nationality,autopct='%1.1f%%',pctdistance=0.9)
axes[1].set_title('2000')

axes[2].pie(nat_year2010.Value,labels=nat_year2010.Nationality,autopct='%1.1f%%',pctdistance=0.9)
axes[2].set_title('2010')

axes[3].pie(nat_year2020.Value,labels=nat_year2020.Nationality,autopct='%1.1f%%',pctdistance=0.9)
axes[3].set_title('2020')

fig.suptitle('Earning distribution',fontsize=20)
fig.tight_layout()
plt.show()
```

Here we can clearly conclude:

USA share is always dominant from 1990s to 2010s of approx more than 78% share of the earning distribution.
but there is a decline in share of USA by 2020 reducing it to around 50%.
